Plot curves on all columns in one graph

```python
import matplotlib.pyplot as plt

# Specify the file path to your local CSV file
file_path = 'C:/Users/acer/Desktop/Assignment/daily-treasury-rates.csv'

# Read the data from the CSV file using pandas
df = pd.read_csv(file_path, header=0, parse_dates=['Date'], index_col='Date')

# Plotting all columns in one graph
plt.figure(figsize=(12, 6))

# Iterate over each column in the DataFrame
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

plt.xlabel('Date')
plt.ylabel('Interest Rates')
plt.title('Interest Rate Curves')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Pairplot all short rates (less than one year rate)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the file path to your local CSV file
file_path = 'C:/Users/acer/Desktop/Assignment/daily-treasury-rates.csv'

# Read the data from the CSV file using pandas
df = pd.read_csv(file_path, header=0, parse_dates=['Date'], index_col='Date')

# Specify the column names for the short rates
short_rate_columns = ['1 Mo','2 Mo', '3 Mo', '4 Mo', '6 Mo']

# Filter the DataFrame to select only the short rates columns
short_rates = df[short_rate_columns]

# Create a pair plot using seaborn
sns.pairplot(short_rates)
plt.show()
```

Scatter plot a linear regression line for 2m and 3m rates

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the file path to your local CSV file
file_path = 'C:/Users/acer/Desktop/Assignment/daily-treasury-rates.csv'

# Read the data from the CSV file using pandas
df = pd.read_csv(file_path, header=0, parse_dates=['Date'], index_col='Date')

# Select the 2-month (2m) and 3-month (3m) rate columns
x = df['2 Mo']
y = df['3 Mo']

# Create a scatter plot with a linear regression line using seaborn
sns.regplot(x=x, y=y, ci=None)

# Add labels and a title
plt.xlabel('2-Month Rate')
plt.ylabel('3-Month Rate')
plt.title('Scatter Plot with Linear Regression Line')

# Display the plot
plt.show()
```

"kde" jointplot 2m and 3m rates.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the file path to your local CSV file
file_path = 'C:/Users/acer/Desktop/Assignment/daily-treasury-rates.csv'

# Read the data from the CSV file using pandas
df = pd.read_csv(file_path, header=0, parse_dates=['Date'], index_col='Date')

# Select the 2-month (2m) and 3-month (3m) rate columns
x = df['2 Mo']
y = df['3 Mo']

# Create a "kde" joint plot for the 2m and 3m rates using seaborn
sns.jointplot(x=x, y=y, kind='kde')

# Add labels and a title
plt.xlabel('2-Month Rate')
plt.ylabel('3-Month Rate')
plt.title('Kernel Density Estimate (KDE) Joint Plot')

# Display the plot
plt.show()
```
