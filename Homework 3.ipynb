```python
import pandas as pd

# Download the CSV file from the provided URL and read it into a DataFrame
url = "https://query1.finance.yahoo.com/v7/finance/download/IBM?period1=1514764800&period2=1546214400&interval=1d&events=history&includeAdjustedClose=true"
df = pd.read_csv(url)

# Add a new column "Trade Value" which is the close price times volume
df['Trade Value'] = df['Close'] * df['Volume']

# Find the highest "Close" price
highest_close_price = df['Close'].max()

# Calculate the average "Close" price
average_close_price = df['Close'].mean()

# Select all rows where the "Close" price is greater than the average "Close" price
selected_rows = df[df['Close'] > average_close_price]

# Convert the "Close" prices to Japanese Yen (JPY) using a fixed exchange rate of 107
df['Close_JPY'] = df['Close'] * 107

# Display the DataFrame
print("IBM 2018 Data:")
print(df)

# Display the highest "Close" price
print("\nHighest Close Price:", highest_close_price)

# Display the average "Close" price
print("Average Close Price:", average_close_price)

# Display selected rows where "Close" price is greater than the average
print("\nSelected Rows:")
print(selected_rows)
```

```python
import pandas as pd
import os

# Define the list of stock symbols
stocks = ['IBM', 'MSFT', 'GOOG', 'AAPL', 'AMZN', 'META']

# Create an empty dictionary to store dividend DataFrames for each stock
dividend_data = {}

# Define a function to compute dividends
def compute_dividends(df):
    df['CloseRatio'] = df['Close'].shift(1) / df['Close']
    df['AdjCloseRatio'] = df['Adj Close'].shift(1) / df['Adj Close']
    df['Dividend'] = (df['CloseRatio'] - df['AdjCloseRatio']) * df['Close']
    dividend_df = df[df['Dividend'].notnull()][['Date', 'Dividend']]
    return dividend_df

# Loop through each stock symbol and read data from files
for stock_symbol in stocks:
    file_path = f"{stock_symbol}.csv"
    
    if os.path.exists(file_path):
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Compute dividends for the stock
        dividend_data[stock_symbol] = compute_dividends(df)
    else:
        print(f"File not found for {stock_symbol}")

# Print the dividend data for each stock
for stock_symbol, df in dividend_data.items():
    print(f"Dividends for {stock_symbol}:")
    print(df)
```
